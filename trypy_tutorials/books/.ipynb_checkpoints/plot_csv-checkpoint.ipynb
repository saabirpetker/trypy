```python
#Let's make a function that does all of the above so we can do it for all our files quickly

def dataframe(filename,delim='\t', headers=['younglings','Test','Shear Rate[1/s]','Shear Stress[Pa]','Shear Strain[%]',
         'Viscosity[mPa·s]','Normal Force[N]','Torque[mN·m]'], ):
    """
    Arguments:
    filename
    
    optional arguments:
    delimiter, headers
    
    Returns:
    clean dataframe of file data
    """
    #get raw
    df = pd.read_csv("../data/ACS2/GR4SM14.csv", encoding="UTF-16", sep=delim, header=None)
    #intro headers
    df.columns=headers
    #kill younglihngs
    df=df.drop(['younglings'],axis=1)
    #clean up
    #Make our new column labels
    a=["loading",]*16
    b=["unloading",]*16
    c=["reloading",]*16
    #for a smaller file
    if len(df)==32:
        d=a+b
    #for a bigger file
    if len(df)==48:
        d=a+b+c
    df['Test']=d
    
    return df
    
    
```

```python
#You guessed it let's make that a function... But I'm going to do something clever
#I'm going to make another function that sets up our axes beforehand 
def initfig(a,b):
    fig, ax = plt.subplots(figsize=(a,b))
    return ax

def split(filename, x, y, ax):
    
    df=dataframe(filename)

    xs=df.loc[:,[x]] 
    ys=df.loc[:,[y]]
    xlab=list(xs.columns) #our labels :)
    ylab=list(ys.columns)

    xs=xs.to_numpy() 
    ys=ys.to_numpy()

    test=df.loc[:,['Test']] 
    test=test.to_numpy() #This is going to be our key:)

    unloadingx=[]
    unloadingy=[]
    loadingx=[]
    loadingy=[]

    for i, ing in enumerate(test):

        if ing == "unloading":
            unloadingx.append(xs[i])
            unloadingy.append(ys[i])

        if ing == "loading":
            loadingx.append(xs[i])
            loadingy.append(ys[i])
    #There's deffo a better way but python is kind.

    ax.scatter(unloadingx,unloadingy,alpha=0.5,marker='o',s=50,label=test[0][0])#I've hardcoded this split here in the label so just be careful
    ax.scatter(loadingx,loadingy,alpha=1,marker='+',s=50,label=test[16][0])#I've hardcoded this split here in the label so just be careful
    ax.legend()
    ax.set_xscale("log")
```

```python
ax1=initfig(5,2)
file="../data/ACS2/GR4SM32.csv"
x1='Shear Rate[1/s]'
y1='Shear Stress[Pa]'
split(file, x, y, ax1)
plt.show()
message()
```

For all the files we care about

```python
#Load all our files with a function
def filelist(path='../data/ACS2', up_to=-1): #I put up_to so we don't have to do all of them every check
    """
    argument:
    path to our data (optional of course)
    returns:
    list of the filenames
    list of titles
    """
    fs=os.listdir(path)
    titles=[]
    for f in fs:
        if f[-4:] =='.csv' and f[:5] =='GR4SM':
            titles.append(f[5:-4])
        
    
    
    return fs[:up_to], titles[:up_to]
#filelist() #check it yourself that it works
```

```python
files, titles = filelist(up_to=-1)

for (file, title) in zip(files,titles):
    ax1=initfig(5,2)
    ax1.set_title(title)
    x1='Shear Rate[1/s]'
    y1='Shear Stress[Pa]'
    split(file, x, y, ax1)
    plt.show()
message()
```
