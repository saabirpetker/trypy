```python
#Let's import our key libraries here...
#These are the toolboxes that hold the tools to solve our problems
import numpy as np
import matplotlib.pyplot as plt
```

#**Quick introduction to functions**
Here we define a function that gives us some output.
We want to make this a mathematical function that takes in a value x and spits out another value, but we won't call it y... yet

```python
def f1(x):
  return 3*x+5
#If I do the following and hit SHIFT+Enter it RETURNS what we expect (if we put a number in x...)
f1(6)
#But if I want to change the function I have to go back into the definition to change anything...
```

```python
#Hold on, there's an easier way to edit my function...
#Let's introduce more variables
def f2(x,m,c):
  return m*x+c
#Now we can generate an arbitrary linear function :)
f2(1,5,6)
```

So far, we have been returning values from cells...
Let's assign these outputs to variables...
In fact, let's make a whole array of these variables

```python
xs=np.linspace(0,100,100)
ys=f2(xs,5,-12)
#What did I do here?
#Python is clever... Instead of inputting a single variable into my function... I put a whole array of variables!
#Then, I assigned this array to my set of y's: ys.
#Now then, maybe I can plot this?
```

Super simple plot

```python
plt.plot(xs,ys)
plt.show() #Always show your plot... This saves you a lot of trouble down the line. For now, just think of it as the return bit of a function
```
