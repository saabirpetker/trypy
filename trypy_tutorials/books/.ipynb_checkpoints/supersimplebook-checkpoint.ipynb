<font size="15">**A Higher Order Plotting Function**

And now the fun part... We can turn the above into a single function itself! This way, we don't need to write a bunch out every time we want to plot something

```python
def simplePlot(func):
    """
    Good practise to leave a text description of functions like this in chonky bois
    This higher order function takes in a function, func and shows us a nice graph
    arguments: func,
    outputs: None
    """
    xs=np.linspace(0,10,10000) #What happens when you change the interval no.?
    ys=func(xs)
    plt.plot(xs,ys)
    plt.show()
```

When we call this with one of our functions we want to plot, we can do so in one line :)


```python
plt.title('plot1')
simplePlot(f1)
#We can plot another function if we want too now! :)
plt.title('plot2') #That's nice, we can inpur title's outside of our functions... remember we hit show() in the function tho...
def newFunc(x):
    x=x+1
    return np.sin(3*np.pi/x)
simplePlot(newFunc)
#Can you try and see what happens with the following?
#simplePlot(f2)
#simplePlot(f3)
```

<font size="15">**Making Things Pretty**

First we should make our plotting function something a bit more... easy... to deal with...
    
If it wasn't already obvious, now it should be... We are very much off-script and into "what's easiest for me?" territory

```python
def prettyPlot(func, save=False, filename=True, quality=100, show=True, title=True, legend=True, axes=True):
    """
    arguments: func,
    outputs: None
    """
    palatinate="#68246D" #This is a surprise tool we will use later
    functionName=func.__name__ #This time let's record the name of our function - the name of any object in python can be recalled in this way :)
    
    #Let's keep what we are plotting the same.
    n=100 #Let's keep track of our maximum to make our title nice:)
    xs=np.linspace(0,n,10000) #What happens when you change the interval no.?
    ys=func(xs)
    #Now when we plot it, however, we won't take shortcuts, which we, unbeknowest to you, have been.
    
    #First, we initialise our figure object, the thing we are chucking things onto
    #As well as the first thing to chuck, the axis, ax1
    
    fig, ax1 = plt.subplots(figsize=(7, 5)) #I've said the size of the figure we are making to 7u by 5u, probs cm, idk.
    
    #Now we say, on the axis, ax1 which is on our figure, let's plot xs and ys. Yay.
    ax1.plot(xs,ys, label="function1",color=palatinate,lw=3,alpha=0.6,) #Mess around with these and checkout matplotlib docs4mor
    
    if axes:#equivalently, if axes==True:
        ax1.set_xlabel("x",fontsize=13)
        ax1.set_ylabel("y",fontsize=13)
    if legend:#equivalently, if legend==True:
        plt.legend()#We can change location of our legend box with bbox arguments - again check documentation
    #Sometimes we don't want titles...
    if title:#equivalently, if title==True:
        title="[0,{}] for function: {}".format(str(n),functionName)
        ax1.set_title(title)
    
    if save:#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html <--Saving is a b-word
        show=True #You have to show things after you try to save things otherwise you won't be saving anything (just working memory things)
        if filename==True: #See python is clever, but sometimes it tries to be too clever, it was printing my error message below without me being explicit... If in doubt, be more explicit ;)
            print("You need to specify your filename. Don't forget your file type extension")
            
        plt.savefig(filename,dpi=quality) # You can change a bunch of things. Mess around.
    if not save:
        print("lovely")
    #Sometimes it's worth having this if statement, just in case you want to make further amendments outside the function
    #However, you should make sure there is a show line somewhere, otherwise you'll run out of cache!!!
    if show: #equivalently, if show==True:
        plt.show()
```

```python
prettyPlot(newFunc) #This should look quite nice 
```

Let's figure out the differences

Let's say we have a list of functions

I'm assuming you've seen a loop before... I might add something on loops at the top actually.. Oh dear.

```python
#Let's make a list of functions to run
func1=f1
func2=newFunc
def func3(x):
    return x**0*np.exp(1) #It's tempting to keep this function as jsut a constant, e.g. return 1, but this won't generate an array in our graph method so we have to be fancy
def func4(x):
    return np.cos(x/4)+np.exp(x)
def func5(x):
    return np.pi/4+10*np.exp(-5/(x+0.0000001))*np.sin(x) #We don't want to divide by zero as well lol

#Let's also make a normalisation function so we make sure everything fits on a [-1,1] range LOL
def norm(ys):
    return ys/np.max(ys)
```

```python
#Let's test a loop that runs through the functions we want to test :)

import sys #I'm cheating a bit here not introducing you to sys before but just go with it, I'm being ott

thismodule = sys.modules[__name__] #This fetches the current module refrence
num_of_funcs=5
for i in range(1, num_of_funcs+1):
    func = getattr(thismodule, "func{}".format(i))
    func(5) #This will invoke func_1 ... func_n
    print(func(5)) #Print check xD
    
#That's how it works in a loop, but let's make this a function haha
def testFunc(funcNum):
    thismodule = sys.modules[__name__] #This fetches the current module refrence
    func = getattr(thismodule, "func{}".format(funcNum))
    return func
#Now let's check the function works:)
print(testFunc(3).__name__)
print('yay')
```

```python
#First let's see how our current functions interact with one another
num_of_funcs=3
for i in range(1,num_of_funcs+1):
    prettyPlot(testFunc(i))
```

First we notice we are dealing with numbers of different scales. Let's say we are only interested inthe graph shape...
1. Let's normalise these functions.
2. We want to put each axis on the same figure
3. We want our legend to differentiate each line
4. Our title should change
5. We want to save-and-show the very final image only

```python
def multiPlot(func, Type='plot', figure=True, save=False, filename=True, quality=100, show=True, title=True, legend=True, axes=True, domain=[[0],[100]], repetitions=1):
    """
    arguments: func,
    outputs: None
    This function attempts to generalise prettyPlot
    """
    palatinate="#68246D" #This is a surprise tool we will use later
    functionName=func.__name__ #This time let's record the name of our function - the name of any object in python can be recalled in this way :)
    
    #Let's keep what we are plotting the same.
    domain[-1]=100 #Let's keep track of our maximum to make our title nice:)
    
    xs=np.linspace(domain[0],domain[-1],10000) #What happens when you change the interval no.?
    if Type=='scatter':
        xs=np.linspace(domain[0],domain[-1],500)
    ys=func(xs)
    ys=norm(ys) #Let's normalise things this time :)
    
    #First, we initialise our figure object, the thing we are chucking things onto
    #As well as the first thing to chuck, the axis, ax1
    if figure:
        fig, ax1 = plt.subplots(figsize=(7, 5)) #I've said the size of the figure we are making to 7u by 5u, probs cm, idk.
    if Type=='plot':
        plt.plot(xs,ys, label=functionName,lw=3,alpha=0.6) #Mess around with these and checkout matplotlib docs4mor
    if Type=='scatter':
        plt.scatter(xs,ys, label=functionName,marker='+',s=5,alpha=1) #Go ham
        
    if axes:
        plt.xlabel("x",fontsize=13)
        plt.ylabel("y",fontsize=13)
    if legend:
        plt.legend(loc="lower left", ncol = repetitions)
    if isinstance(title,str):
        plt.title(title)
        
    if save:#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html <--Saving is a b-word
        show=True #You have to show things after you try to save things otherwise you won't be saving anything (just working memory things)
        if filename==True: #See python is clever, but sometimes it tries to be too clever, it was printing my error message below without me being explicit... If in doubt, be more explicit ;)
            print("You need to specify your filename. Don't forget your file type extension")
            
        plt.savefig(filename,dpi=quality) # You can change a bunch of things. Mess around.
    if not save:
        print("lovely")
    #Sometimes it's worth having this if statement, just in case you want to make further amendments outside the function
    #However, you should make sure there is a show line somewhere, otherwise you'll run out of cache!!!
    if show: #equivalently, if show==True:
        plt.show()
```

```python
num_of_funcs=5
#Things to do for the first axis
domain=[[0],[300]]
figure=True
show=False
axes=False
legend=False
save=False
quality=100
filename="my_first_multiplot.png"
title="Comparing shape of my functions: \n"
repetitions=1
for i in range(1,num_of_funcs+1):
    title=title+testFunc(i).__name__+", "
    #print(title)
    repetitions+=1
    multiPlot(testFunc(i), figure=figure, save=save, filename=filename, quality=quality, show=show, title=title, legend=legend, axes=axes, repetitions=repetitions, )
    #Things to do after the first axis
    figure=False
    if i==num_of_funcs-1: #Things to do before the last axis
        show=True
        axes=True
        legend=True
        title=title
        #save=True
        #quality=100
```
